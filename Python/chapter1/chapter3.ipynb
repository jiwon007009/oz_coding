##내장함수 : divmod

> 내장함수 : 자주 사용하는 기능을 미리 만들어 놓은 것

divmod : 몫과 나머지를 동시에 구해줌

```python
print(divmod(6, 2))
print(divmod(7, 4))
print(divmod(13, 2))
print(divmod(19, 4))
```

```python
print(7 // 4 , 7 % 4)
```

### 괄호

수학에서 사용되는 괄호와 동일한 기능

() : 괄호 안에 있는 것을 먼저 계산

사칙연산 순서 : 괄호 > 곱나눔 > 덧,뺄셈

```python
3 + 2 * 4
```

```python
(3+ 2)* 4
```

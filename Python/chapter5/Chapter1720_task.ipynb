__[문제7]__

오즈 (oz) 클래스에서 이름, 나이, 성별을 출력하는 who() 메소드를 추가하고 who()메소드를 이용해 아래 출력 결과를 만들어주세요


출력 결과 ) <br>
이름: 김코, 나이: 30, 코딩레벨: A

```python
class oz(): # 클래스의 인스턴스 생성
    def __init__(self, name,age, coding_level): # 기본 생성자
        self.name = name
        self.age = age
        self.coding_level = coding_level

    def who(self):
        print(f'이름: {self.name}, 나이: {self.age}, 코딩레벨: {self.coding_level}')

coding = oz('김코' , 30,'A')
coding.who()
```

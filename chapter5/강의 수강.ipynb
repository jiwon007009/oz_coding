```python
class Person:
    def __init__(self): #self는 인스턴스 자기 자신을 의미
        self.hello = '안녕하세요'

    def greeting(self):
        print(self.hello)

호출 = Person()#클래스 무조건 선언하고 메쏘드 진행
호출.greeting() # 메쏘드
```

```python
class Person:
    def __init__(self, name, age, address): #self는 인스턴스 자기 자신을 의미
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.address = address

    def greeting(self):
        print(f'{self.hello} 저는 {self.name}입니다.')

maria = Person('마리아',20, '서울시 서초구 반포동') # 1. 인스턴스 생성 -> 먼저 클래스에 매개변수 할당
maria.greeting() # 2.메쏘드 호출

print('이름 :', maria.name)
print('나이 :', maria.age)
print('주소 :', maria.address)
```
